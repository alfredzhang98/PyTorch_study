过拟合和欠拟合的概念实际上就是模型的表现效果。接下来，我们再来看看损失函数和代价函数，这组概念就是我们刚才说用来衡量“偏差”、“效果”的方法。我们还是延续之前的思路，用函数举例子。
假设刚才的二维空间中，任意一个点对应的真实函数为 F(x)。我们通过模型的学习拟合出来的函数为 f(x)。
根据刚才提到的学习过程，我们会知道 F(x) 和 f(x) 之间存在一个误差，我们定义为 L(x)：

$L(x) = L(F(x), f(x))$

## 常见损失函数

### 0-1 损失函数（0-1 Loss Function）
最简单的 0-1 损失函数，这个函数的公式表示如下：

$L(F(x), f(x)) = \begin{cases} 0, & \text{if } F(x) = f(x) \\ 1, & \text{if } F(x) \neq f(x) \end{cases}$

F(x) 是输入数据的真实类别，f(x) 是模型预测的类别

### 平方损失函数（Squared Loss Function）
$L(x) = (F(x) - f(x))^2$

这里 F(x) 和 f(x) 的差距我们做了一个平方和，是为了保证两者的误差是一个正值，方便后续的计算。
梯度更新，那时你就会发现，平方和要比绝对值更为方便

我们会在损失函数中加入一个 1/2 的系数，这是为了求导的时候能够跟平方项的系数约掉。


### 均方差损失函数和平均绝对误差损失函数
$MSE = \frac{1}{n} \sum_{i=1}^{n} (F(x_i) - f(x_i))^2$

$MAE = \frac{1}{n} \sum_{i=1}^{n} |F(x_i) - f(x_i)|$

### 交叉熵损失函数（Cross-Entropy Loss Function）

信息熵
$H(p) = - \sum_{i=1}^{C} p(x_i) \log(p(x_i))$

p(x_i) 是类别 i 的概率，C 是类别总数


$CE = - \sum_{i=1}^{C} p(x_i) \log(q(x_i))$

p(x_i) 是真实类别的概率分布，q(x_i) 是模型预测的概率分布


### softmax 损失函数（Softmax Loss Function）
$Softmax(j) = \frac{e^{a_j}}{\sum_{k=1}^{C} e^{a_k}}$

这里 a_j 是类别 j 的得分，C 是类别总数

带入回交叉熵损失函数中：

$ \sum_{i=1}^{C} p(x_i) \log(Softmax(j)) $


softmax with cross-entropy loss，它是交叉熵损失函数的一个特例




损失函数的值越小越好么？

是的，损失函数的值越小，表示模型的预测结果与真实结果之间的差距越小，模型的表现越好。因此，在训练过程中，我们通常会通过优化算法（如梯度下降）来最小化损失函数的值，从而提升模型的性能。
但是，有时候损失函数的值虽然很小，但模型在测试集上的表现并不好，这可能是由于过拟合导致的。因此，除了关注损失函数的值外，我们还需要综合考虑模型在训练集和测试集上的表现，以确保模型具有良好的泛化能力。
